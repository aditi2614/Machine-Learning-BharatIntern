```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
df = pd.read_csv('housing3.csv')
x = df['LSTAT']
y = df['MEDV']
x = x.to_numpy()
x = x.reshape(-1,1)
regression_model = LinearRegression()
regression_model.fit(x, y)
y_predicted = regression_model.predict(x)
print(y_predicted)
```

```python
import numpy as np
N = np.alen(y)
y_sum = np.sum(y)
y_pr_sum = np.sum(y_predicted)
print(y_sum - y_pr_sum)
```

```python
from sklearn.metrics import mean_absolute_error
print("MAE = ",mean_absolute_error(y, y_predicted))
```

```python
from sklearn.metrics import mean_squared_error
print("MSE = ",mean_squared_error(y, y_predicted))
```

```python
from sklearn.metrics import r2_score
print("R_Squared = ",r2_score(y, y_predicted))
```

```python
from sklearn.metrics import r2_score
r_sqr = r2_score(y, y_predicted)
N = len(y)
k = 1
print("Adjusted R_Squared = ",(1-r_sqr)*(N-1)/(N-k-1))
```
